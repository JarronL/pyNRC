```python
from astropy.convolution import convolve_fft
from scipy import fftpack
from time import sleep
from memory_profiler import memory_usage
from astropy import units as u
from astropy.utils.console import human_file_size

def f(fov_pix=32, **kwargs):
    arr = np.zeros([fov_pix,fov_pix])
    kern = np.zeros([fov_pix*2,fov_pix*2])
    im_conv = convolve_fft(arr, kern, fftn=fftpack.fftn, ifftn=fftpack.ifftn, allow_huge=True)
    
    del arr, kern, im_conv
    
def f0(pause=5):
    sleep(pause)
```

```python
array = np.asarray(np.zeros([4024,4024]), dtype=np.complex)
arrayshape = array.shape
array_size_B = (np.product(arrayshape, dtype=np.int64) *
                np.dtype(np.complex).itemsize)*u.byte
print("Arrays will be %s." % human_file_size(array_size_B.to(u.byte).value))
```

```python
# Direct Imaging Calculations
fov_arr = np.array([16,32,128,160,256,320,512,640,1024,2048])

mem_vals = [] #np.zeros([fov_arr.size])
for i,fov_pix in enumerate(fov_arr):
    print(fov_pix)
    mem_max0 = memory_usage((f0,(),{'pause':2}), max_usage=True, interval=0.01)
    mem_max = memory_usage((f,(),{'fov_pix':fov_pix}), max_usage=True, interval=0.01)
    mem_vals.append(mem_max[0]-mem_max0[0])
    
mem_vals = np.array(mem_vals)
```
